### Simple linear Regression Example

Linear Regression Formula
The simplest regression is a straight line:

ùë¶ =ùëö‚ãÖùë• + c

ùë¶: predicted value (e.g., house price)

ùë•: input feature (e.g., square footage)

ùëö: slope (how much y changes when x increases)

c : intercept (baseline value when x = 0)

## Steps to Implement Simple Linear Regression

```python
# import libraries 
import pandas as pd
from sklearn.linear_model import LinearRegression

# import warnings 
import warnings
warnings.filterwarnings('ignore')  # Suppress all warnings

# Sample data
data = pd.DataFrame({
    'SquareFeet': [1000, 1500, 2000, 2500, 3000],
    'Price': [200000, 250000, 300000, 350000, 400000]
})

X = data[['SquareFeet']]   # Feature
y = data['Price']          # Target

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict price for a 2200 sq ft house
predicted_price = model.predict([[2200]])
print(predicted_price)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Fit model
model = LinearRegression()
model.fit(X, y)

# Parameters
m = model.coef_[0]        # slope
b = model.intercept_      # intercept

# Predictions and cost
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)

print(f"slope (m): {m:.2f}, intercept (b): {b:.2f}")
print(f"MSE (cost): {mse:.2f}")

# Plot
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.grid(True)
plt.text(1, max(y)-0.5, f'MSE (Cost): {mse:.2f}', fontsize=12, color='black')
plt.show()
```
